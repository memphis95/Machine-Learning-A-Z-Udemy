```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Position_Salaries.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

#### 2. Splitting the dataset into train and test

### 3. Model Training

#### Training the Linear Regression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
```

```python
lin_reg.fit(X, y)
```

#### Training the Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
# poly_feature = PolynomialFeatures(degree=2)
# poly_feature = PolynomialFeatures(degree=3)
poly_feature = PolynomialFeatures(degree=4)
X_poly = poly_feature.fit_transform(X)
```

```python
lin_reg_poly = LinearRegression()
lin_reg_poly.fit(X_poly, y)
```

### 4. Visualization


#### Linear Regression (Base Models)

```python
plt.scatter(X,y, color="red")
plt.plot(X, lin_reg.predict(X), color="orange")
plt.title("Base Model (Linear Regression)")
plt.xlabel("Position Index")
plt.ylabel("Salaries")
```

#### Polynomial Regression (degree = 2)

```python
plt.scatter(X,y, color="red")
plt.plot(X, lin_reg_poly.predict(X_poly), color="pink")
plt.title("Base Model (Linear Regression)")
plt.xlabel("Position Index")
plt.ylabel("Salaries")
```

#### Polynomial Regression (degree = 3)

```python
plt.scatter(X,y, color="red")
plt.plot(X, lin_reg_poly.predict(X_poly), color="pink")
plt.title("Base Model (Linear Regression)")
plt.xlabel("Position Index")
plt.ylabel("Salaries")
```

#### Poylnomial Regression (degree = 4)

```python
plt.scatter(X,y, color="red")
plt.plot(X, lin_reg_poly.predict(X_poly), color="pink")
plt.title("Base Model (Linear Regression)")
plt.xlabel("Position Index")
plt.ylabel("Salaries")
```

#### Smoothing the polynomial curve

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X,y, color="red")
plt.plot(X_grid, lin_reg_poly.predict(poly_feature.fit_transform(X_grid)), color="pink")
plt.title("Base Model (Linear Regression)")
plt.xlabel("Position Index")
plt.ylabel("Salaries")
```
